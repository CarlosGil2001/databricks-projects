##### Métodos de Normalización de Datos

###### 1. Escalado de características Simple
f1.png

Divide cada uno por el valor máximo. Entre 0 y 1

###### 2. Min-Max
f2.png

Toma cada valor y le resta el mínimo, luego se divide por el rango de esa característica. Entre 0 y 1

###### 3. Puntaje Z
f3.png

Resta Mu que es el promedio y lo divide por la desviación estándar. Entre -3 a +3, pueden ser más altos o bajos.

```python
# Utilizarémos la normalización a las características "length", "width" y "height"
# Aplicando el método "ESCALADO DE CARACTERÍSTICAS SIMPLE"
data["lenght"]=data["lenght"]/data["lenght"].max()
```

f4.png

```python
# Aplicando el método "MIN-MAX"
data["lenght"]= (data["lenght"] - data["lenght"].min()) / (data["lenght"].max() - data["lenght"].min())
```

f5.png

```python
# Aplicando el método "Puntaje Z"
data["lenght"] = (data["lenght"] - data["lenght"].mean()) / data["lenght"].std()
```

f6.png

```python
# Haríamos lo mismo con las características "width" y "height"
```

## Regresión de Polinomios y Canalización (Pipelines)
- ¿Qué hacemos cuando un modelo lineal no se ajusta mejor a nuestros datos? <br>
Utilizamos: **Regresión de Polinomios** <br>
- Transformamos nuestros datos en un polinomio y luego usamos la regresión lineal para ajustar el modelo. <br>
<br>
**Canalización:** Son una forma de simplificar el código.
**Regresión Polinomial:** Es un caso especial de la regresión lineal, describe relaciones curvilíneas. <br>
<br>
- **¿Qué es una relación Curvilínea?**<br>
Es lo que se obtiene elevando al cuadrado o de orden superior de las variables predictoras en el modelo.


**2nd Orden (Elevado al cuadrado)**
- Lo que significa que la variable predictora del modelo está elevado al cuadrado.

- Lo que significa que la variable predictora del modelo está elevado al cubo. <br>

<br>

```python
# Desarrollar una base de modelo de regresión polinomial de 2nd orden (variable predictora al cuadrado)

x=data['highway-mpg'] # variable predictora
y=data['price'] # variable objetivo

# Calcular poligono de 2nd orden
f=np.polyfit(x,y,2)
p=np.poly1d(f)
print(p)
```

```python
# Desarrollar una base de modelo de regresión polinomial de 3rd orden (variable predictora al cubo)

x=data['highway-mpg'] # variable predictora
y=data['price'] # variable objetivo

# Calcular poligono de 3rd orden
f=np.polyfit(x,y,3)
p=np.poly1d(f)
print(p)
```

### Canalización (Pipelines) <br>
- Podemos simplificar nuestro código con una librería de **canalización**.
- Hay muchas formas de obtener una predicción, mediante:
###### **Normalización, transformación polinomial y regresión lineal.** <br>

image.png

```python
# Importamos los módulos que necesitamos
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Creamos una lista de tuplas
# El primer elemento de la tupla contiene el nombre del estimador: modelo.
# El segundo elemento contiene el constructor de modelos.
Input=[('scale', StandardScaler()),('polynomial', PolynomialFeatures(degree=2)),('model', LinearRegression())]
```

```python
# Ingresamos la lista en el constructor de canalización
pipe=Pipeline(Input)
pipe
```

```python
# Entrenar canalización
x_predict= data[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
pipe.fit(x_predict,y)

# Predecir
Yhat_pipe=pipe.predict(x_predict[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']])
Yhat_pipe
```

image.png

```python
# Gráfico de distribución

# Compara el valor predecido frente al valor real.
# Ideal para modelos con más de una variable independiente (predictora)

ax1= sns.distplot(y, hist=False, color="r", label="Actual Value")
sns.distplot(Yhat_pipe, hist=False, color="b", label="Fitted Values", ax=ax1)

#Podemos observar que la precisión es casi exacta
```

## Evaluación del Modelo
- Nos dice cómo funciona nuestro modelo en el mundo real.
- **Se debe dividir nuestros datos, usar los datos In-sample o los datos de entrenamiento para entrenar el modelo**
- **El resto de los datos, denominados datos de prueba, se utilizan fuera de la muestra**
- Los **datos de prueba** se utilizan para aproximar cómo se comporta el modelo en el mundo real.
- Separar los datos en conjuntos de entrenamiento y prueba es una parte importante de la evaluación.

### Al dividir conjunto de datos:
- Mayor parte de los datos se usan para entrenamiento, por ejemplo: **70%**
- Una menor parte para prueba, por ejemplo: **30%**

### Pasos:
- Usamos el conjunto de entrenamiento para construir un modelo y descubrir relaciones predictivas.
- Luego usamos el conjunto de prueba para evaluar el rendimiento del modelo.

```python
# Necesitamos evaluar el modelo con datos nuevos que el modelo no haya visto antes.
# Dividir conjunto de datos con la función "train_test_split()"
```

### Error Generalizado
- Es una medida de qué tan bien nuestros datos predicen cosas nunca antes vistas.
- **El error que obtenemos usando nuestros datos de prueba es una aproximación de este error.**

```python
# GRÁFICO DE REGRESIÓN - TRAIN

# Comparar el valor actual de los precios con los valores predichos (train).

ax1= sns.distplot(y_data, hist=False, color="r", label="Actual Value")
sns.distplot(y_train, hist=False, color="b", label="Predicted Values", ax=ax1)
```

```python
# GRÁFICO DE REGRESIÓN - TEST

# Comparar el valor actual de los precios con los valores predichos de prueba (test).

ax1= sns.distplot(y_data, hist=False, color="r", label="Actual Value")
sns.distplot(y_test, hist=False, color="b", label="Predicted Values", ax=ax1)
```

- Al generar la misma gráfica y comparando los datos de entrenamiento y prueba, vemos que los datos de prueba son diferentes.
- **La diferencia se debe a un error generalizado y represente lo que vemos en la realidad.**
- Para solucionar este problema utilizamos la **Validación Cruzada.**

#### Validación Cruzada
- En este método, el conjunto de datos se divide en **k-grupos iguales,** se hace referencia a cada grupo como un pliegue.

#### Por ejemplo: **4 pliegues**
image.png

- Podemos usar **3 pliegues para entrenar**, luego use **un pliegue para prueba**, esto se repite hasta que cada partición de use tanto para entrenamiento como para prueba.
- Al final, usamos los resultados promedio como la estimación del error fuera de la muestra.

```python
# La métrica de evaluación depende del modelo.
# Por ejemplo: R-cuadrado.
```

```python
# La forma más sencilla de aplicar la VALIDACIÓN CRUZADA es llamar a la función "cross_val_score" que realiza múltiples evaluaciones 'fuera de la muestra'
```

```python
# Importamos librería necesaria
from sklearn.model_selection import cross_val_score
```

```python
# Usamos función "cross_val_score"

lr=LinearRegression() # lr: tipo de modelo
x_data= data[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']] # variables predictoras
y_data=data["price"] # variable objetivo
# cv: número de particiones o divisiones

# Calculamos la VALIDACIÓN CRUZADA
scores=cross_val_score(lr,x_data,y_data,cv=3)
scores
```

```python
# Podemos promediar el resultado juntos para estimar R-cuadrado 
np.mean(scores)
```

```python
# Saber la predicción real, antes que se calculen los valores R-cuadrado
from sklearn.model_selection import cross_val_predict
yhat=cross_val_predict(lr,x_data,y_data,cv=3)
yhat
```

```python
# Evaluación con GRÁFICO DE REGRESIÓN
ax1= sns.distplot(y_data, hist=False, color="r", label="Actual Value")
sns.distplot(yhat, hist=False, color="b", label="Predicted Values", ax=ax1)
```
