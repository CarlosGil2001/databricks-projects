### Agregar un valor predeterminado o completar los datos que faltan

```python
#Cambiar por un valor predeterminado a los valores null
df_with_0=data.fillna(0) #Guardamos en una variable la data con los valores NULL reemplazados por 0. 
```

```python
#Mostramos la data
df_with_0.head()
```

```python
# LA MEJOR MANERA DE LLENAR LOS VALORES NULL, ES CON UNA MEDIDA

# Llenarlo con la media(mean), en la columna con valores NULL
data['DURATION'].mean() #verificamos la media
```

```python
#Llenamos los valores NULL con la media(mean) en la columna DURATION
data_with_mean=data.DURATION.fillna(data['DURATION'].mean())
```

```python
#Verificamos los valores
data_with_mean
```

### Dropping Duplicates - Eliminar duplicados

- drop_duplicates() <br>
- keep='first' <br>

```python
#Verificar si hay duplicados (coincidencia en todos los valores)
data.duplicated()
```

```python
#Especificar si hay valores duplicados por una columna en especÃ­fico
data.duplicated('MOVIE_TITLE') #Duplicados con el mismo titulo de pelÃ­cula
```

```python
#Eliminamos los duplicados en la columna 'MOVIE_TITLE'
data_drop_dup=data.drop_duplicates('MOVIE_TITLE')
```

```python
#Verificamos si hay duplicados por MOVIE_TITLE
data_drop_dup.duplicated('MOVIE_TITLE')
#Ya no existe duplicados
```
